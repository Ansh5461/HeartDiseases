```python
y = heart["HeartDisease"]
x = heart.drop(['HeartDisease'], axis = 1, inplace = True)
```

```python
y
```

```python
x = heart
x
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.3, random_state = 1234, shuffle=True)
```

```python
print("X_train.shape ", x_train.shape)
print("X_test.shape", x_test.shape)
print("y_train.shape", y_train.shape)
print("y_test.shape", y_test.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix

rfc = RandomForestClassifier()

rfc.fit(x_train, y_train)

y_pred = rfc.predict(x_test)

print("Training score ", rfc.score(x_train, y_train))
print("Testing accuracy ", rfc.score(x_test, y_test))
print("Confusion matrix \n", confusion_matrix(y_test, y_pred))
```

```python
from sklearn.tree import DecisionTreeClassifier
dtc=DecisionTreeClassifier(criterion='entropy', random_state=0)

dtc.fit(x_train, y_train)

y_pred = dtc.predict(x_test)

print("Training score ", dtc.score(x_train, y_train))
print("Testing accuracy ", dtc.score(x_test, y_test))
print("Confusion matrix \n", confusion_matrix(y_test, y_pred))
```

```python
from sklearn.linear_model import Perceptron

perceptron = Perceptron(max_iter=5)

perceptron.fit(x_train, y_train)

y_pred = perceptron.predict(x_test)

print("Training score ", perceptron.score(x_train, y_train))
print("Testing accuracy ", perceptron.score(x_test, y_test))
print("Confusion matrix \n", confusion_matrix(y_test, y_pred))
```
